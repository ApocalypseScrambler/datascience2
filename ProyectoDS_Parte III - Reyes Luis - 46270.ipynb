<h3 style="text-align:left;border-radius:8px;font-family:Times;font-variant:small-caps;color:#FFFFFF;background-color:#3300CC;">Information Value</h3>

*Análicemos las variables Gender, Marital_Status, Income_Category, Card_Category y Education_Level, para saber si pueden darnos mas información sobre nuestra variable target.*

```python
# Genero un dataframe que voy a ir completando con cada IV
IV_df = pd.DataFrame(columns=['Variable', 'IV'])
IV_df.head()
```

```python
# Calcula el Information Value (IV) para la variable "Gender"
tabla_IV = df.groupby('Gender').agg(      
    Cant = ('Attrition_Flag', 'count'),
    Cant_Exito = ('Attrition_Flag', 'sum'),    
    Tasa_Exito = ('Attrition_Flag', 'mean')
).reset_index()

tabla_IV
```

*Debido a un error que no pudimos determinar, no podemos calcular correctamente el logaritmo de nuestros datos. Para solucionarlo, pasamos los datos anteriores a un nuevo dataframe para ahi si realizar los datos.*

<h3 style="text-align:left;border-radius:8px;font-family:Times;font-variant:small-caps;color:#FFFFFF;background-color:#3300CC;">K-fold Cross-Validation - Validación Cruzada 1</h3>

```python
from sklearn.model_selection import StratifiedKFold
# Crear un modelo de regresión logística
clf = LogisticRegression()

# Definir el número de pliegues (k-fold)
n_splits = 5  # Puedes ajustar esto según tus necesidades

# Crear un objeto StratifiedKFold para dividir los datos
kf = StratifiedKFold(n_splits=n_splits, random_state=42, shuffle=True)

# Inicializar listas para almacenar las métricas de precisión en cada pliegue
accuracy_scores = []

# Realizar la validación cruzada
for train_index, test_index in kf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Entrenar el modelo en el conjunto de entrenamiento
    clf.fit(X_train, y_train)

    # Predecir en el conjunto de prueba
    y_pred = clf.predict(X_test)

    # Calcular la precisión en el conjunto de prueba
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Imprimir las métricas de precisión en cada pliegue
for i, accuracy in enumerate(accuracy_scores):
    print("Precisión en el pliegue {}: {:.2f}%".format(i + 1, accuracy * 100))

# Calcular la precisión promedio
mean_accuracy = sum(accuracy_scores) / n_splits
print("Precisión promedio en validación cruzada: {:.2f}%".format(mean_accuracy * 100))
```

*Al realizar esta validación cruzada hemos corroborado la generalidad de nuestro modelo, como a su vez que no tenemos un problema de subajuste o sobreajuste.*
