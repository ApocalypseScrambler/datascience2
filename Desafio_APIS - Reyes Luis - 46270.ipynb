```python
# Importamos Librerias a usar
import requests
import pandas as pd
```

*Documentación de la API:*
*https://rickandmortyapi.com/documentation/*

```python
# Función para obtener los datos de un personaje específico por su ID
def obtener_datos_personaje(personaje_id):
    url = f'https://rickandmortyapi.com/api/character/{personaje_id}'
    response = requests.get(url)
    
    if response.status_code == 200:
        # Devolver los datos del personaje en formato JSON
        return response.json()
    else:
        # Imprimir un mensaje de error si la solicitud no fue exitosa
        print(f"Error al obtener datos del personaje {personaje_id}: {response.status_code}")
        return None

# Lista para almacenar los datos de todos los personajes
personajes_data = []

# Iterar sobre los IDs de los personajes (1 a 826)
for personaje_id in range(1, 827):
    data = obtener_datos_personaje(personaje_id)
    
    if data:
        personajes_data.append(data)

# Crear un DataFrame con los datos de todos los personajes
df = pd.DataFrame(personajes_data)

# Extraer el nombre de la ubicación y el origen de cada personaje
df['origin'] = df['origin'].apply(lambda x: x['name'] if 'name' in x else None)
df['location'] = df['location'].apply(lambda x: x['name'] if 'name' in x else None)

# Mostrar el DataFrame
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurar el estilo de seaborn
sns.set(style="whitegrid")

# Calcular la distribución de las categorías de status
status_counts = df['status'].value_counts()

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Cantidad de Personajes por Estados', fontsize=16)
plt.axis('equal')  # Proporciona un círculo perfecto

plt.show()
```

```python
plt.figure(figsize=(18, 6))

cantidad = len(df['species'])

sns.countplot(data = df,
              x = "species", palette="viridis")
plt.xlabel('Especie', fontsize=14)
plt.ylabel('Cantidad', fontsize=14)
plt.title('Cantidad de Personajes por Especie', fontsize=16)
ax = plt.gca()
for p in ax.patches:
    x_text = p.get_x() + p.get_width() / 2.0  # Calcula la posición x del texto
    y_text = p.get_height()  # Toma la altura de la columna
    ax.text(x_text, y_text, f'{str(round(int(y_text)/cantidad,2) * 100)}%', # Indica posición x e y, y texto a mostrar 
            ha = 'center', # Alineación horizontal 
            va = 'bottom',  # Alineación vertical
            fontsize = 10, color = 'black', fontweight = 'bold')
plt.show()
```

```python
from wordcloud import WordCloud

def WC(col,super=True,by=None,top=100): # a wordcloud function that takes a categorical column, then..
    t = ''
    if not by: # if you didn't  give it a numric column to size by it's values, it'll..
        for item in df[col].dropna(): # loop over the main categorical column, and..
            if super: # checks if that main column is a column of lists, it'll..
                for i in item: # loop over each item in each list in the column, then it..
                    if len(i) > 1: # checks if that item is more than 1 character string, it..
                        t += f' {i}' # adds that word to a comulative text.
            else: # and if the main column wasn't a column of lists, it..
                t += f' {item}' # directly adds each item in the column in the comulative text, but..
    else: # if you did give it a numric column to size by it's values, it'll..
        for item in df.sort_values(by=by,ascending=False)[col].dropna().head(top): # it loops only on the top movies based on that given numric column, and it
            if super: # checks if that main column is a column of lists, it'll..
                for i in item: # loop over each item in each list in the column, then it..
                    if len(i) > 1: # checks if that item is more than 1 character string, it..
                        t += f' {i}'# adds that word to a comulative text.
            else:  # and if the main column wasn't a column of lists, it..
                t += f' {item}' # directly adds each item in the column in the comulative text, but..

    plt.figure(figsize=(12, 8), dpi=100) # after all of that it creates a figure to..

    word_cloud = WordCloud(collocations = False, width=800, height=400, background_color = 'white').generate(t) # plot on a word cloud of each word in the comulative text.

    plt.imshow(word_cloud, interpolation='bilinear') 

    plt.axis("off")

    plt.show()
```
